```python
def compare_dfs(df1,df2):
    print(df1.shape)
    print(df2.shape)
    if set(df1.columns.values) == set(df2.columns.values):
        print("Same Feature names")
    elif df1.columns.values != df2.columns.values:
        print("Different feature names")
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
```

```python

def process_missing_values(input_df,strat="most_frequent"):
    imputer = SimpleImputer(strategy=strat)
    imputed_df = pd.DataFrame(imputer.fit_transform(input_df))
    imputed_df.columns = input_df.columns
    imputed_df.index = input_df.index
    #preserve data types
    imputed_df = imputed_df.astype(input_df.dtypes.to_dict())
    return imputed_df

imputed_train_df = process_missing_values(X_train_full)
imputed_test_df = process_missing_values(X_test_full)

```

```python
def get_numerical_columns(input_df):
    return [ col for col in input_df.columns.values
                       if input_df[col].dtype == "float64" or \
                         input_df[col].dtype == "int64"]

def get_categorical_columns(input_df):
    return [col for col in input_df.columns.values
            if input_df[col].dtype == "object"]

ordinal_columns = ["ExterQual", "ExterCond", "BsmtQual", "BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2",\
"HeatingQC","KitchenQual","Functional","FireplaceQu","GarageQual","GarageCond"]

# Note: Ensure matching category values. Ensure consistent categories for train and test data.
# So that can succesfully ordinal and one-hot encode.
def valid_categorical_columns(candidate_columns,train_df,test_df):
    valid_columns = []
    for col in candidate_columns:
        # Note: maybe sets should be equal?
        if set(test_df[col]).issubset(set(train_df[col]) ) :
            valid_columns.append(col)
    return valid_columns

train_numerical_cols = get_numerical_columns(imputed_train_df)
test_numerical_cols = get_numerical_columns(imputed_test_df)

train_categorical_columns = get_categorical_columns(imputed_train_df)
test_categorical_columns = get_categorical_columns(imputed_test_df)

#Note: check if not columns missing after split
print(len(imputed_train_df.columns), len(imputed_test_df.columns))
print(len(train_numerical_cols), len(test_numerical_cols))
print(len(train_categorical_columns),len(test_categorical_columns))

print(set(train_categorical_columns) == set(test_categorical_columns))

valid_ordinal_cols = valid_categorical_columns(ordinal_columns,imputed_train_df,imputed_test_df)


```

```python
def ordinal_encode_df(input_df):
    ordinal_encoder = OrdinalEncoder()
    encoded_df = pd.DataFrame(ordinal_encoder.fit_transform(input_df))
    encoded_df.index = input_df.index
    encoded_df.columns = input_df.columns
    return encoded_df
    

ordinal_train_df = ordinal_encode_df(imputed_train_df[valid_ordinal_cols])
ordinal_test_df = ordinal_encode_df(imputed_test_df[valid_ordinal_cols])

```

```python
def get_low_high_cardinality_columns(input_df,column_list, max_cardinality):
    low_card = []
    high_card = []
    for col in column_list:
        if input_df[col].nunique() <= max_cardinality and input_df[col].dtype == "object":
            low_card.append(col)
        else:
            high_card.append(col)
    return low_card, high_card
```

```python
# Note: Can I reduce the cardinality of high cardinality columns by using an "Other" label for low frequency values
# Based on: https://towardsdatascience.com/dealing-with-features-that-have-high-cardinality-1c9212d7ff1b

from collections import Counter
def reduce_cardinality(column, threshold = 0.75, return_categories = True):
    threshold_value = threshold*len(column)
    frequency_sum = 0

    new_category_list = []
    counts=Counter(column)
    most_common = dict(counts.most_common())

    for category,count in most_common.items():
        frequency_sum = frequency_sum + count
        new_category_list.append(category)
        if frequency_sum >= threshold_value:
            break
    # Test
    new_category_list.append("Other")
    new_column = column.apply(lambda x: x if x in new_category_list else "Other")
    
    if return_categories:
        return new_column,new_category_list
    else:
        return new_column


# transformed_col, new_categories = reduce_cardinality(imputed_X_train["Neighborhood"])
```

```python
# Note: before one-hot encoding ....
# 1. get columns to one hot encode
one_hot_cols = [col for col in train_categorical_columns if col not in valid_ordinal_cols]
valid_one_hot_cols = valid_categorical_columns(one_hot_cols,imputed_train_df,imputed_test_df)
# 2. get low/high cardinality
# Note: same result when ran on imputed_test_df... therefore, high_card columns are same between train and test
low_card, high_card = get_low_high_cardinality_columns(imputed_train_df,valid_one_hot_cols, 8)
print(len(low_card),len(high_card))
# 3. reduce cardinality
OH_train_df = imputed_train_df[low_card]
OH_test_df = imputed_test_df[low_card]

for col in high_card:
    transformed_train_data_col = reduce_cardinality(imputed_train_df[col], threshold=0.8, return_categories=False)
    OH_train_df = pd.concat([OH_train_df,transformed_train_data_col.to_frame()],axis=1)

    transformed_test_data_col = reduce_cardinality(imputed_test_df[col], threshold=0.8, return_categories=False)
    OH_test_df = pd.concat([OH_test_df,transformed_test_data_col.to_frame()],axis=1)
```

```python
def combine_dfs(numerical_df, ordinal_df, one_hot_df):
    dfs_to_combine = [numerical_df,ordinal_df,one_hot_df]
    output_df = pd.concat(dfs_to_combine,axis=1)
    return output_df


X_train_final_df = combine_dfs(imputed_train_df[train_numerical_cols],ordinal_train_df,one_hot_encoded_train_df)
X_test_final_df = combine_dfs(imputed_test_df[test_numerical_cols],ordinal_test_df,one_hot_encoded_test_df)

```
